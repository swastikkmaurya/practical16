```python
#1. Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
text = "Natural Language Processing is Interesting"
tokens = word_tokenize(text)
print(tokens)
```

```python
from nltk.tokenize import sent_tokenize
text = "Python is easy. It is widely used in data science."

sentences = sent_tokenize(text)
print(sentences)
```

```python
#Stop word removal
#Stop words such as is the , and , in are removed

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

words = word_tokenize(text)
filtered_words = [word for word in words if word.lower() not in stop_words]
print(filtered_words)
```

```python
#stemming
#stemming reduces words to thier root form
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["running", "ran", "run"]
stemmed_words = [stemmer.stem(word) for word in words]
words = ["python", "pythoner", "pythoning", "pythoned"]
for word in words:
    print(stemmer.stem(word))
```

```python
#lemmatization
#lemmatization converts words to thier dictionary form
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize("Writing"))
print(lemmatizer.lemmatize("Studies"))
print(lemmatizer.lemmatize("Best"))
print(lemmatizer.lemmatize("running"))
print(lemmatizer.lemmatize("Coding"))
```

```python
#part- of - Speech (POS) tagging
#POS tagging indentifies grammatical roles of words.
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
words = word_tokenize("Python is a powerful language for programming")
pos_tags = pos_tag(words)
print(pos_tags)
```

```python
#word Frequency Count
from nltk.probability import FreqDist
words = word_tokenize(text)
fd = FreqDist(words)
print(fd.most_common())
```
